```python
def convert_to_instruction_format(dataset):
  formatted_data = []
  for idx, row in tqdm(dataset.iterrows()): # Thêm index vào trong dữ liệu
    formatted_data.append({
      "Intruction": "Tóm tắt văn bản sau:",
      "input": f'Tiêu đề: {row["Title"]}\n\nNội dung: {row["Contents"]}',
      "output": row["Summary"]
    })
  return formatted_data
```

```python
# save file jsonl
def save_jsonl(formatted_data, file_name):
  with open(f'{file_name}.jsonl', 'w', encoding='utf-8') as f:
    for sample in formatted_data:
      f.write(json.dumps(sample, ensure_ascii=False) + '\n')
```

```python
def convert_for_all(list_dataset):
  for file_name, dataset in list_dataset.items():
    formatted_data = convert_to_instruction_format(dataset)
    save_jsonl(formatted_data, file_name)
```

```python
list_dataset = {
    'train': dataset_train,
    'test': dataset_test,
    'val': dataset_val
}

convert_for_all(list_dataset)
```
